```python
#fill the missing values using any one method
column_name='age'
df[column_name]=df[column_name].fillna(df[column_name].mean())
print(df)
```

```python
#visualise the data set using two appropriate plots
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.hist(df['avg_glucose_level'],bins=10)
plt.title('Brain Stroke')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```

```python
#use any two out of three normalization methods to normalize two columns in the given data
from sklearn.preprocessing import MinMaxScaler,StandardScaler
columns_to_normalize=["avg_glucose_level","bmi"]
min_max_scaler=MinMaxScaler()
df_min_max=df.copy()
df_min_max[columns_to_normalize]=min_max_scaler.fit_transform(df[columns_to_normalize])
print("Min_Max_scaling:")
print(df_min_max[columns_to_normalize].head())
```

```python
z_score_scaler=StandardScaler()
df_z_score=df.copy()
df_z_score[columns_to_normalize]=z_score_scaler.fit_transform(df[columns_to_normalize])
print("z_score_scaling:")
print(df_z_score[columns_to_normalize].head())
```
